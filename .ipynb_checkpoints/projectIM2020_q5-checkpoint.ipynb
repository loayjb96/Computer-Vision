```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
# import the necessary packages
from skimage.measure import compare_ssim
path1 = glob.glob("Q2DB/*.png")
path2 = glob.glob("Q4DB/*.jpg")
Image_DB_Q3 = []
Image_DB_Q4=[]


def Read_img(path_to_directory,db):
    for img in path_to_directory:
        n = cv2.imread(img)
        db.append(n)
def Detect_and_compare():
    img_to_show=[]
    name_in_db=[]
   
    sift = cv2.xfeatures2d.SIFT_create()
    found=0
    j=0
    for img1 in Image_DB_Q3:
        sharpen_kernel = np.array([[-1,-1,-1], [-1,10,-1], [-1,-1,-1]])
#         sharpen = cv2.filter2D(img1, -1, sharpen_kernel)
        fig=plt.figure(figsize=(5, 5))
        print('Checking '+path1[j]+'  against database')
        plt.imshow(Image_DB_Q3[j])
        plt.show()
        k=0
        j+=1
        i=0
        kp_1, desc_1 = sift.detectAndCompute(img1, None)
        for img2 in Image_DB_Q4:
                i+=1

                img2=cv2.GaussianBlur(img2,(5,5),0)
#                 sharpen1 = cv2.filter2D(img2, -1, sharpen_kernel)
                kp_2, desc_2 = sift.detectAndCompute(img2, None)
                index_params = dict(algorithm=0, trees=5)
#                 index_params= dict(algorithm = 6,table_number = 6,key_size = 12,multi_probe_level = 1) #2
                search_params = dict(checks=50)
                flann = cv2.FlannBasedMatcher(index_params, search_params)        
                matches = flann.knnMatch(desc_1, desc_2, k=2)
                good_points = []
                ratio = 0.8
                for m, n in matches:
                    if m.distance < ratio*n.distance:
                        good_points.append(m)
                if(len(good_points)>90):
                    img_to_show.append(Image_DB_Q4[k])
                    name_in_db.append(path2[k])
                    print(len(good_points),path2[k])
                matches=0  
                k+=1
                if(i==len(Image_DB_Q4)):
                    if(len(img_to_show)==0):
                        print("No match Found!!! \n")
                    else:   
                        print("Displaying the Images \n")
                    showImages(img_to_show,name_in_db)
                    img_to_show=[]
                    name_in_db=[]
        
def showImages(ImgDB,path):

    fig=plt.figure(figsize=(20, 50))
    columns = 5
    rows = 10
   
    for i in range(1,len(ImgDB)+1):
        fig.add_subplot(rows, columns, i)
        plt.imshow(ImgDB[i-1])
        plt.xticks([]), plt.yticks([])
        plt.title(path[i-1])
    plt.show()                    
Read_img(path1,Image_DB_Q3)
Read_img(path2,Image_DB_Q4)
Detect_and_compare()
```
