```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
# import the necessary packages
from skimage.measure import compare_ssim
path1 = glob.glob("Q3Images/*.png")
path2 = glob.glob("Q2DB/*.png")
cv_img = []
Image_DB=[]
image_set1=[]
image_set2=[]
max_found_value=[]

 
def Read_img(path_to_directory,db,read):
    for img in path_to_directory:
        n = cv2.imread(img)
        if(read==1):
            n=cv2.fastNlMeansDenoisingColored(n,None,10,10,7,21)
            
        
        db.append(n) 

    
def showImages(ImgDB):
    w=10
    h=10
    fig=plt.figure(figsize=(10, 10))
    columns = 4
    rows = 5
   
    for i in range(1,len(ImgDB)+1):
        fig.add_subplot(rows, columns, i)
        plt.imshow(ImgDB[i-1])
        plt.title(path1[i-1])
    plt.show()
def Detect_and_compare():
   
    sift = cv2.xfeatures2d.SIFT_create()
    found=0
    j=0
    for img1 in cv_img:
        
#         n=cv2.medianBlur(img1, 5);
#         n = cv2.cvtColor(n, cv2.COLOR_BGR2GRAY)
   
        fig=plt.figure(figsize=(5, 5))
        print('Checking '+path1[j]+'  against database')
        plt.imshow(cv_img[j])
        plt.show()
        k=0
        j+=1
        i=1
        kp_1, desc_1 = sift.detectAndCompute(img1, None)
        for img2 in Image_DB:
#                 n=cv2.medianBlur(img2, 5);
#                 n = cv2.cvtColor(n, cv2.COLOR_BGR2GRAY)
#                 n=cv2.fastNlMeansDenoisingColored(img2,None,10,10,7,21)
                i+=1
                kp_2, desc_2 = sift.detectAndCompute(img2, None)
                index_params = dict(algorithm=2, trees=5)
#                 index_params= dict(algorithm = 6,table_number = 6,key_size = 12,multi_probe_level = 1) #2
                search_params = dict(checks=60)
                flann = cv2.FlannBasedMatcher(index_params, search_params)
                
                matches = flann.knnMatch(desc_1, desc_2, k=2)
                good_points = []
                ratio = 0.7
                for m, n in matches:
                    if m.distance < ratio*n.distance:
                        good_points.append(m)
                
#                 if(j==3):
                    
                max_found_value.append(len(good_points))
#                 print(len(good_points))
                if(len(good_points)>30):
                    print(len(good_points))
                    found=1
                    plt.title('match found to  '+path2[k]) 
                    fig.add_subplot(5,4, i)
                    i+=1
                    plt.imshow(Image_DB[k])
                    plt.show()
                matches=0  
                k+=1
                if(found==1):
                    found=0
                    break 
                if(i==len(Image_DB) and found==0):
                    print(max(max_found_value))
                    a=max_found_value.index(max(max_found_value))
                    max_found_value[a]=0
                    a=max_found_value.index(max(max_found_value))
                    print(max_found_value)
                    fig=plt.figure(figsize=(5, 5))
                    
                    plt.imshow(Image_DB[a])
                    plt.show()

                    
Read_img(path1,cv_img,1)
Read_img(path2,Image_DB,0)
Detect_and_compare()
showImages(cv_img)
```
