```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
path = glob.glob("Q2Images/*.png")
cv_img = []
for img in path:
    n = cv2.imread(img)
    cv_img.append(n)  
def concatenate(img1,img2):
    h1, w1 = img1.shape[:2]
    h2, w2 = img2.shape[:2]
#create empty matrix
    vis = np.zeros((max(h1, h2), w1+w2,3), np.uint8)

#combine 2 images
    vis[:h1, :w1,:3] = img1
    vis[:h2, w1:w1+w2,:3] = img2
    return vis
def show(img):    
    cv2.imshow('Output for question 2', img) 
# De-allocate any associated memory usage  
    if cv2.waitKey(0): 
        cv2.destroyAllWindows() 
def showImages(cv_img):
    vis1 = np.concatenate((cv_img[0],cv_img[1],cv_img[2],cv_img[3],cv_img[4]), axis=1)
    vis2 = np.concatenate((cv_img[5],cv_img[6],cv_img[7],cv_img[8],cv_img[9]), axis=1)
    vis1 = cv2.resize(vis1, (650, 300), None, .12, .15)
    vis2 = cv2.resize(vis2, (650, 300), None, .12, .15)
    visfinal = np.concatenate((vis1,vis2), axis=0)     
    cv2.imshow('Output for question 2', visfinal) 
# De-allocate any associated memory usage  
    if cv2.waitKey(0): 
        cv2.destroyAllWindows() 
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2,2))

def Preformjob(cv_img):
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2,3))
    Copy_cv_img = []
    i=0
    changeable=0
    maxRad=0
    for img in cv_img:
        n = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        n=cv2.morphologyEx(n, cv2.MORPH_CLOSE, kernel)
#         show(n)
        i+=1
        if(i==len(cv_img)-1 or i==len(cv_img)-3):
            changeable=17
            if i==len(cv_img)-1:    
                maxRad=32
            else:
                maxRad=15
        else:
            changeable=63
            maxRad=55
        circles=cv2.HoughCircles(n, cv2.HOUGH_GRADIENT, 1, minDist=22,param1=400,param2=changeable,minRadius=5,
               maxRadius=maxRad)    
        if circles is not None:
            circles = np.round(circles[0, :]).astype("int")
            for (x, y, r) in circles:
                cv2.circle(img, (x, y), r, (105, 255, 0), 4)
                cv2.rectangle(img, (x - 5, y - 5), (x + 5, y + 5), (0, 128, 255), -1)  
        Copy_cv_img.append(n)
Preformjob(cv_img)
showImages(cv_img)
```
